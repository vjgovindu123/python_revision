```python
import pandas as pd
data=pd.read_csv("E:\DataSets\Mckinsey.csv")
data
print(data.shape)
print(data.columns)
print(data.keys)
print(data[['country','gdp_cap']].tail())
print("The continents are",data['continent'].unique())
print(data['continent'].value_counts())
print(data.rename({"population":"Population","country":"Country"},axis=1,inplace=True))
print("continents are")
print(data.columns)
data.drop('continent',axis=1,inplace=True)#Drop complete continent column with axis=1(Column-Wise)
data
data["year+7"]=data["year"]+7
print(data.head())
```

```python
#Calculate Gross GDP from above dataset
data["gdp"]=data["Population"]*data["gdp_cap"]
print(data.head())
```

```python
#Remove newly added values
data.drop(columns=["year+7","gdp"],axis=1,inplace=True)
```

```python
movies_data=pd.read_csv("E:\DataSets\movies.csv")
directors_data=pd.read_csv("E:\DataSets\directors.csv")
print(movies_data)
directors_data
movies_data.drop(columns=["Unnamed: 0"],inplace=True)
print(movies_data)
directors_data.drop(columns=["Unnamed: 0"],inplace=True)
print(directors_data)
```

```python
print(directors_data.head())
print("Movies Data")
print(movies_data.head())
print(movies_data["year"].max())
print(movies_data["year"].min())
print("-- Years of Movies_Data is given",movies_data["year"].max()-movies_data["year"].min()+1)
print("Describe Movies Data",movies_data.describe)
```

```python
#Forward String using func
movies_data["Released-Day"]=movies_data.apply(lambda movies_data:f"{movies_data['day']}-{movies_data['month']}-{movies_data['year']}",axis=1)
print(movies_data.head())
#or using concatenation
#movies_data["Released_date"]=movies_data['day'] + movies_data['month']+ movies_data['year']--Type Conversion Error
movies_data["Released_date"]=movies_data['day']+movies_data['month']+movies_data['year'].astype('str')#or apply(str)
print(movies_data['Released_date'])
movies_data.drop(['Released-Day'],axis=1,inplace=True)
print(movies_data)
```
